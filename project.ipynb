```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder , LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataforEncoding.csv')
```

```python
df
```

```python
df['Disease'].value_counts()
```

```python
X = df.iloc[: , [1 ,2 ,3 ,4 ,5 ,6 ,7 ,8]]
Y = df['Outcome Variable']
```

```python
X.head()
```

```python
#thing is that We Apply OneHotEncoding and Ordinal Encoding
```

```python
df.isnull().sum()
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , Y , test_size = 0.2)
```

```python
X_train
```

```python
df['Blood Pressure'].value_counts()
df['Cholesterol Level'].value_counts()
```

```python
oe = OrdinalEncoder(categories = [['Low' , 'Normal' , 'High']])
```

```python
X_train_bp = oe.fit_transform(X_train[['Blood Pressure']])
X_test_bp = oe.transform(X_test[['Blood Pressure']])
```

```python
X_train_cl = oe.fit_transform(X_train[['Cholesterol Level']])
X_test_cl = oe.transform(X_test[['Cholesterol Level']])
```

```python
#now we will do Nominal Encoding to remaining all inpute set except Age
```

```python
oh = OneHotEncoder(drop = 'first' , sparse_output = False)
```

```python
X_train_f = oh.fit_transform(X_train[['Fever']])
X_test_f = oh.fit_transform(X_test[['Fever']])
```

```python
X_train_c = oh.fit_transform(X_train[['Cough']])
X_test_c = oh.fit_transform(X_test[['Cough']])

X_train_ft = oh.fit_transform(X_train[['Fatigue']])
X_test_ft = oh.fit_transform(X_test[['Fatigue']])

X_train_db = oh.fit_transform(X_train[['Difficulty Breathing']])
X_test_db = oh.fit_transform(X_test[['Difficulty Breathing']])

X_train_g = oh.fit_transform(X_train[['Gender']])
X_test_g = oh.fit_transform(X_test[['Gender']])
```

```python
X_train_f
```

```python
X_train['isFever'] = X_train_f
X_test['isFever'] = X_test_f

X_train['isCough'] = X_train_c
X_test['isCough'] = X_test_c

X_train['isFatigue'] = X_train_ft
X_test['isFatigue'] = X_test_ft

X_train['isDifficultyBreathing'] = X_train_db
X_test['isDifficultyBreathing'] = X_test_db

X_train['isMale'] = X_train_g
X_test['isMale'] = X_test_g

X_train['BPLevel'] = X_train_bp
X_test['BPLevel'] = X_test_bp

X_train['CL_Level'] = X_train_cl
X_test['CL_Level'] = X_test_cl
```

```python
X_test
```

```python
X_train.drop(['Fever' , 'Cough' ,'Fatigue'  ,'Difficulty Breathing', 'Gender' , 'Cholesterol Level', 'Blood Pressure'] , axis = 1 , inplace = True)
```

```python
X_test.drop(['Fever' , 'Cough' ,'Fatigue'  ,'Difficulty Breathing', 'Gender' , 'Cholesterol Level', 'Blood Pressure'] , axis = 1 , inplace = True)
```

```python
le = LabelEncoder()
```

```python
y_train_new = le.fit_transform(y_train)
```

```python
y_test_new = le.transform(y_test)
```

```python
y_train_new = pd.DataFrame(y_train_new , columns = ['Outcome'])
y_test_new = pd.DataFrame(y_test_new , columns = ['Outcome'])
```

```python
y_train_new
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train_age = sc.fit_transform(X_train[['Age']])
X_test_age = sc.transform(X_test[['Age']])
```

```python
X_train['Age'] = X_train_age
X_test['Age'] = X_test_age
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train , y_train_new.values.ravel())
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test_new , y_pred)
```
